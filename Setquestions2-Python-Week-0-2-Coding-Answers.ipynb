```python
"""4.Write a Python program to count the number of vowels in a given string. 
Expected Output (for input "education"):

Number of vowels: 5"""
```

```python
vowels=['a','e','i','o','u']
str1=input("Enter the string:")
str2=str1.lower()
count=0
for i in vowels:
    count+=str2.count(i)
print("Number of vowels: ",count)
```
