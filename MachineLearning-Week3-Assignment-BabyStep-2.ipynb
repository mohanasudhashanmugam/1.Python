```python
#importing the library pandas to handle database and files related functions
import pandas as pd
```

```python
#read_csv function is called to read csv file and assign it to dataset variable
dataset=pd.read_csv("Salary_Data.csv")
dataset
```

```python
#Spliting input and output by assigning them to i/p & o/p variables respectively

independent=dataset[["YearsExperience"]]
dependent=dataset[["Salary"]]
independent
```

```python
# spliting training set and test set data seperately using train_test_split function from sklearn library 
# which supports all machine learning functions
# Arguments to the function - independent variable, dependent variable and the test data ratio=30%
 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30)
```

```python
#Model creation - Learning phase
#Import Linear Regression class from sklearn.linear_model

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()     #creating object "regressor" for LinearRegression() class
regressor.fit(x_train,y_train)   #Passing training data i/p and o/p to fit function to find weight and 
                                    #bais. Once w,b is found then model gets created

weight=regressor.coef_           # assigning the calculated slope value from coef_ variable to weight variable
bais=regressor.intercept_        #  assigning the calculated bais value from intercept_ variable to bais variable




```

```python
#Evaluation Metric phase

y_pred=regressor.predict(x_test) #predict method is called to calculate "predicted o/p" by passing x_test as input

# to cross check the model is performing better or not, r2_score function is imported. It calculates r2 value.
# If r2 value is nearer or equal to 0, the model is performing poor
# If r2 value is nearer or equal to 1, the model is performing better

from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score # r2value - 0.9646133098205328 is nearly to 1, so we conclude model's performance is good
```

```python
# Save model phase
import pickle  # importing pickle module which supports model saving
filename="Finalised_model_linear.sav" # Assigning a filename to a variable
pickle.dump(regressor,open(filename,'wb')) #using dump function, open a file in write mode 
                                                #and write the model into the file *.sav and save
```

```python
#Load the model to cross check the result before deployment
load_model=pickle.load(open("Finalised_model_linear.sav",'rb')) #Model is loaded and opened in read mode
result=load_model.predict([[15]]) #input is passed to predict the output
```
