```python
#4.Write a Python function to find the factorial of 5. Expected Output:
#120

```

```python
def factorial(n):
    if n==0 or n==1 :
        return 1
    elif n>0:
        fact=n*factorial(n-1)  
        return fact
    
```

```python
print(factorial(5))
```
