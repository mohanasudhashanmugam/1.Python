```python
##Body Mass Index
```

```python
BMI=int(input("Enter the BMI index:"))
if(BMI<18.5):
    print("Underweight")
elif(BMI>=18.5 and BMI<=24.9):
    print("Healthyweight")
elif(BMI>=25 and BMI<=29.9):
    print("Overweight")
else:
    print("Very Overweight")
```
